```python
import random
spinner=random.randint(1,100)
def spin(spinner):
    if spinner>0 and spinner<51:
        return "Loss"
    elif spinner>50 and spinner<100:
        return "Win"
    elif spinner==100:
        return "Draw"
print spin(spinner)
print(spinner)
```
